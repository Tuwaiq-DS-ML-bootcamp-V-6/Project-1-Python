# Database where student data stored

```python
students = {}
with open('students.txt', 'r') as file:
    for line in file:
        fields = line.strip().split(', ')
        entry = {
            'id': int(fields[0].split(': ')[1]),
            'name': fields[1].split(': ')[1],
            'gpa': float(fields[2].split(': ')[1]),
            'hours': int(fields[3].split(': ')[1]),
            'final': int(fields[4].split(': ')[1]),
            'hw': int(fields[5].split(': ')[1]),
            'quiz': int(fields[6].split(': ')[1]),
            'lab': int(fields[7].split(': ')[1]),
            'grade': fields[8].split(': ')[1]
            }
        students[entry['id']] = entry
```

# System Functions
View, Search, Filter, Add, Remove, Edit

```python
# Update - to update values added, removed, or edited
def update_students():
    with open('students.txt', 'w') as txt_file:
        for student_id, student_info in students.items():
            txt_file.write("Student ID: {}, ".format(student_id))
            txt_file.write("Name: {}, ".format(student_info['name']))
            txt_file.write("GPA: {}, ".format(student_info['gpa']))
            txt_file.write("Hours: {}, ".format(student_info['hours']))
            txt_file.write("Final: {}, ".format(student_info['final']))
            txt_file.write("hw: {}, ".format(student_info['hw']))
            txt_file.write("Quiz: {}, ".format(student_info['quiz']))
            txt_file.write("Lab: {}, ".format(student_info['lab']))
            txt_file.write("Grade: {}, ".format(student_info['grade']))
            txt_file.write("\n")
```

```python
# View Students
def view_students(students, indent = 0):
    for key, value in students.items():
        if isinstance(value, dict):
            print(str(key) + ":" + " " * indent)
            view_students(value, indent + 1)
        else:
            print(str(key) + ": " + str(value))
    print("-----------------------------------")
```

```python
def search_students(id):
    id = int(id)
    value = students.get(id, students[id]) 
    return value if value else f"ID {id} not found"
```

```python
# Filter Students
def filter_students():
    filter_criteria = {}

    print("Choose one filter criteria:")
    print("1. Standing (freshman, sophomore, junior, senior)")
    print("2. GPA threshold")
    print("3. Pass/Fail status (Pass or Fail)")

    choice = input("Enter your choice (1, 2, or 3): ")

    if choice == '1' or choice == '2' or choice == '3':
        if choice == '1':
            standing = input("Enter standing (freshman, sophomore, junior, senior): ").lower()
            if standing:
                filter_criteria['standing'] = standing
        if choice == '2':
            gpa_threshold = input("Enter GPA threshold: ")
            if gpa_threshold:
                filter_criteria['GPA'] = float(gpa_threshold)
        if choice == '3':
            pass_fail = input("Enter Pass/Fail status (Pass or Fail): ")
            if pass_fail:
                filter_criteria['pass_fail'] = pass_fail
    else:
        print("Invalid choice.")
        return

    filtered_students = []
    for student_id, student_info in students.items():
        if 'standing' in filter_criteria:
            standing = filter_criteria['standing']
            if standing == 'freshman':
                if student_info['hours'] <= 30:
                    filtered_students.append({student_id: student_info})
            elif standing == 'sophomore':
                if 30 < student_info['hours'] <= 60:
                    filtered_students.append({student_id: student_info})
            elif standing == 'junior':
                if 60 < student_info['hours'] <= 90:
                    filtered_students.append({student_id: student_info})
            elif standing == 'senior':
                if student_info['hours'] > 90:
                    filtered_students.append({student_id: student_info})

        if 'GPA' in filter_criteria:
            gpa_threshold = filter_criteria['GPA']
            if student_info['gpa'] >= gpa_threshold:
                filtered_students.append({student_id: student_info})

        if 'pass_fail' in filter_criteria:
            pass_fail = filter_criteria['pass_fail'].lower()
            if pass_fail == 'pass':
                if student_info['grade'][:4] == 'Pass':
                    filtered_students.append({student_id: student_info})
            elif pass_fail == 'fail':
                if student_info['grade'][:4] == 'Fail':
                    filtered_students.append({student_id: student_info})
    
    # print output
    print("----------------")
    for n in filtered_students:
        n = dict(n)
        student = n[list(n.keys())[0]]
        if "standing" in filter_criteria:
            print('ID: %s | Name: %s | Hours: %d' % (student["id"], student["name"], student["hours"]))
        elif "GPA" in filter_criteria:
            print('ID: %s | Name: %s | GPA: %d' % (student["id"], student["name"], student["gpa"]))
        elif "pass_fail" in filter_criteria:
            print('ID: %s | Name: %s | Grade: %s' % (student["id"], student["name"], student["grade"]))
            print("Home Works: %d | Quizzes: %d | Lab: %d | Final: %d" % (student["hw"], student["quiz"], student["lab"], student["final"]))

    
```

```python
# Add
def generate_id():
    ids = list(students.keys())
    if not ids:
        return 1
    ids = iter(ids)
    id = int(next(ids))
    while(True):  
        try:
            if id + 1 == int(next(ids)):
                id += 1
            else:
                return id + 1
        except:
            return id + 1

def checkResults(final, HW, Quiz, Lab):
    try:
        inputs = [int(n) for n in [final, HW, Quiz, Lab]]
    except:
        print("values should be integers")
    result = sum(inputs)
    return "Pass" if result > 59 else "Fail"
    
def add_element():
    student_info = {}
    student_id = generate_id()
    try:
        student_info['id'] = student_id
        student_info['name'] = input("Enter student name: ")
        student_info['gpa'] = float(input("Enter student GPA: "))
        student_info['hours'] = int(input("Enter student hours: "))
        student_info['final'] = int(input("Enter student final grade: "))
        student_info['hw'] = int(input("Enter student HW grade: "))
        student_info['quiz'] = int(input("Enter student Quiz grade: "))
        student_info['lab'] = int(input("Enter student Lab grade: "))
        student_info['grade'] = checkResults(student_info['final'], student_info['hw'], student_info['quiz'], student_info['lab'])
    except:
        print("Enter the correct input for the required field.")
    students[student_id] = student_info
    update_students()
    print(f"Student with ID {student_id} added successfully.")
```

```python
# Remove
def remove_element(id):
    if id in students:
        del students[id]
        update_students()
        print(f"Student with id {id} removed successfully.")
    else:
        print(f"There are no student with the ID '{id}'.")
```

```python
# Edit
def edit_student_info(student_id):
    student_id = int(student_id)
    if student_id in students:
        student_info = dict(students[student_id])
        field = input('Enter the field to edit: ').lower()
        if not student_info.get(field):
            print("Field not found.")
            return
        
        if field == 'id':
            print("Cannot edit ID.")
            return
        
        if field == 'grade':
            print("Cannot edit Grade")

        new_value = input('Enter the new value: ')
        
        student_info[field] = new_value
        student_info['grade'] = checkResults(student_info['final'], student_info['hw'], student_info['quiz'], (student_info['lab']))
        students[student_id] = student_info
        update_students()
        print(f"Field [{field}] updated to {new_value} for student ID {student_id} successfully.")
        return 
    else:
        print("Student ID not found.")
        return
```

# System Menu

```python
def main():
    
    print('''Student Management System
    1- View Students
    2- Search Student
    3- Filter Students
    4- Add Student
    5- Remove Student
    6- Edit Student
    7- Exit
    ''')
    while(True):
        user_input = input("Choose Your Option\n>> ")
        if user_input == "1":
            view_students(students)
        elif user_input == "2":
            user_input = input("Enter Student's ID")
            print(search_students(user_input))
            pass
        elif user_input == "3":
            filter_students()
        elif user_input == "4":
            add_element()
        elif user_input == "5":
            user_input = int(input('Enter Student ID: '))
            remove_element(user_input)
        elif user_input == "6":
            edit_student_info(input('Enter Student ID: '))
        elif user_input == "7":
            update_students()
            break
    
    print("Thank you for using the Student Management System. Goodbye!")
main()
```
